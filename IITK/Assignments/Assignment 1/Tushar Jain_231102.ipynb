```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import det
```

```python
# Q1
array_3d = np.random.randint(1, 101, (4, 4, 4))
mean = np.mean(array_3d)
std_dev = np.std(array_3d)
z_score_normalized = (array_3d - mean) / std_dev
print("3D Array:")
print(array_3d)
print("Z-score Normalized Array:")
print(z_score_normalized)
```

```python
# Q2
iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', 
                   header=None, 
                   names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
filtered_iris = iris[iris['petal_length'] > 3.0]
print("Filtered Iris Data:")
print(filtered_iris)
```

```python
# Q3: Plot a line graph
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='y = sin(x)', color='blue')
plt.title('Line Graph of y = sin(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
# Q4: Create a DataFrame and perform operations
data = pd.DataFrame({
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': [np.nan if i % 2 == 0 else np.random.rand() for i in range(10)]
})
data['C'].fillna(data['C'].mean(), inplace=True)
sorted_data = data.sort_values(by=['A', 'B'], ascending=[True, False])
print("Original DataFrame:")
print(data)
print("Sorted DataFrame:")
print(sorted_data)
```

```python
# Q5: Calculate dot product and determinant
matrix1 = np.random.randint(1, 10, (3, 3))
matrix2 = np.random.randint(1, 10, (3, 3))
dot_product = np.dot(matrix1, matrix2)
determinant = det(dot_product)
print("Matrix 1:")
print(matrix1)
print("Matrix 2:")
print(matrix2)
print("Dot Product:")
print(dot_product)
print("Determinant:")
print(determinant)
```

```python
# Q6: Group DataFrame by a categorical column
categories = ['A', 'B', 'C']
grouped_data = pd.DataFrame({
    'Category': np.random.choice(categories, 20),
    'Values': np.random.randint(1, 100, 20)
})
group_sum = grouped_data.groupby('Category').sum()
print("Grouped DataFrame:")
print(group_sum)
```

```python
# Q7: Create a histogram
numerical_data = np.random.randn(1000)
plt.figure(figsize=(8, 5))
plt.hist(numerical_data, bins=20, color='purple', edgecolor='black')
plt.title('Histogram of Numerical Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Q8: Perform slicing on a 2D array
array_2d = np.random.randint(1, 20, (6, 6))
sub_array = array_2d[1:4, 2:5]
print("2D Array:")
print(array_2d)
print("Sliced Sub-array:")
print(sub_array)
```

```python
# Q9: Merge two DataFrames
data1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Value1': ['A', 'B', 'C', 'D']
})
data2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Value2': ['X', 'Y', 'Z', 'W']
})
merged_data = pd.merge(data1, data2, on='ID', how='outer')
print("Merged DataFrame:")
print(merged_data)
```

```python
# Q10: Create a scatter plot
data_scatter = pd.DataFrame({
    'X': np.random.rand(50),
    'Y': np.random.rand(50)
})
plt.figure(figsize=(8, 5))
plt.scatter(data_scatter['X'], data_scatter['Y'], color='green', marker='*', label='Data Points')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
