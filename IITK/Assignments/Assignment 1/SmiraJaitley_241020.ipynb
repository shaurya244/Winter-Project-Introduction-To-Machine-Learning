```python
import numpy as np
array = np.random.randint(1, 101, size=(4, 4, 4))

mean = array.mean()
std_dev = array.std()

z_score_normalized = (array - mean) / std_dev
print("Original Array:\n", array)
print("\nZ-Score Normalized Array:\n", z_score_normalized)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)
data = {"Column1": np.random.randint(1, 100, 10),
        "Column2": np.random.rand(10) * 100,
        "Column3": np.random.randint(1, 100, 10).astype(float)}

data["Column3"][4] = np.nan

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
df.fillna(df.mean(), inplace=True)
print("\nDataFrame after replacing NaN values with column mean:")
print(df)

sorted_df = df.sort_values(by=["Column1", "Column2"], ascending=[True, False])
print("\nDataFrame after sorting by 'Column1' and 'Column2':")
print(sorted_df)
```

```python
matrix_a = np.random.randint(1, 10, (3, 3))
matrix_b = np.random.randint(1, 10, (3, 3))

print("Matrix A:")
print(matrix_a)
print("\nMatrix B:")
print(matrix_b)

result_matrix = np.dot(matrix_a, matrix_b)
print("\nDot product of Matrix A and Matrix B:")
print(result_matrix)

determinant = np.linalg.det(result_matrix)
print("\nDeterminant of the resulting matrix:")
print(determinant)
```

Q9

```python
data1 = {
    "ID": [1, 2, 3, 4],
    "Name": ["English", "Civics", "Math", "Chem"],
    "Score": [85, 90, 78, 92]
}
data2 = {
    "ID": [3, 4, 5, 6],
    "Subject": ["Physics", "Biology", "History", "Art"],
    "Grade": ["A", "B", "C", "A"]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print("\nDataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

# Merge the DataFrames on the common column 'ID'
merged_df = pd.merge(df1, df2, on="ID", how="inner")
print("\nMerged DataFrame:")
print(merged_df)
```
