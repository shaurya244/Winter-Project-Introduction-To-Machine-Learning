Q1. Using NumPy, create a 3D array of shape (4, 4, 4) filled with random numbers between 1 and 100. Find the z score normalisation of each element of the array.

```python
import numpy as np
arr = np.random.randint(1,100,(4,4,4))
print('Original Data :\n',arr)
print('-'*50)

#Z-score normalization of the array
mean = arr.mean()
std = arr.std()
Z_scores = (arr - mean)/std
print('Z Score Normalized Data :\n', Z_scores)
```

Q2. Using Pandas, filter the rows in the Iris dataset where the petal length is greater than 3.0, and display the filtered data.

```python
import pandas as pd
from sklearn import datasets

#Loading the iris dataset
iris = datasets.load_iris()
df = pd.DataFrame(data = iris.data, columns = iris.feature_names)

#Obtaining the required data
data = df.loc[df['petal length (cm)']>3.0]
print('Data Obtained:\n',data)
```

Q4. Create a Pandas DataFrame with random values with at least 3 columns and perform the following operations: (i) Replace NaN values with the mean of the column. (ii) Sort the DataFrame based on any two columns.

```python
#Generating a dataset with NaN
arr = np.random.rand(6,3)
arr[0,0] = np.nan
arr[4,1] = np.nan
arr[3,2] = np.nan
df = pd.DataFrame(arr)
print('Original Data:\n', df)
print('-'*30)

#Replacing NaN with the mean of the column
mean = df.mean()
new_df = df.replace(np.nan,mean)
print('New Data:\n',new_df)
print('-'*30)

#Sorting the new dataframe
sorted_df = new_df.sort_values(by=[1,2], ascending=[True,True])
print('Sorted Data:\n',sorted_df)
```

Q5. Using NumPy, calculate the dot product of two matrices and find the determinant of the resulting matrix.

```python
matrix1 = np.random.randint(1,20,(4,4))
matrix2 = np.random.randint(1,20,(4,4))

#Obtaining the product of 2 matrices
product = matrix1@matrix2
print('Matrix 1:\n',matrix1)
print('Matrix 2:\n', matrix2, '\n')
print('Product of Matrices:\n', product)
print('-'*30)

#Obtaining the determinant
determinant = np.linalg.det(product)
print('Determinant of Matrix:\n',determinant)
```

Q6. With Pandas, group a DataFrame by a categorical column and calculate the sum of another numerical column for each group.

```python
df1 = pd.read_csv('shopping_trends.csv')
#Obtaining the dataset
df2 = df1.iloc[1:21,[4,5]]

#Grouping the categories and calculating the sum
result = df2.groupby('Category').sum()['Purchase Amount (USD)']
print(result)
```

Q9. Using Pandas, perform a merge operation between two DataFrames based on a common column for dataset of your choice.

```python
data = pd.read_csv('shopping_trends.csv')
#dataset 1
df1 = data.iloc[0:5,[0,3]]
print('Data Frame 1:\n',df1)
print('-'*30)

#dataset 2
df2 = data.iloc[0:5,[0,6]]
print('Data Frame 2:\n',df2)
print('-'*30)

#Merging based on Customer ID
df3 = pd.merge(df1,df2, on = 'Customer ID')
print('Merged Data Frame:\n\n',df3)
```
