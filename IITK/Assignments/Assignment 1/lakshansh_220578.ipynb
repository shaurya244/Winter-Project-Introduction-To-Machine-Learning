```python
# Q1
import numpy as np

arr3d = np.random.randint(1, 101, (4, 4, 4))

# Calculate the mean and standard deviation of the array
mean = np.mean(arr3d)
std = np.std(arr3d)

# Perform z-score normalization
z_score_normalized_array = (arr3d - mean) / std

arr3d, z_score_normalized_array
```

```python
#Q2
import pandas as pd

# Load the Iris dataset from a built-in source
from sklearn.datasets import load_iris

# Load the dataset into a DataFrame
iris_data = load_iris()
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)

# Add the target column for species
iris_df['species'] = iris_data.target

# Filter rows where petal length (column index 2) is greater than 3.0
filtered_data = iris_df[iris_df.iloc[:, 2] > 3.0]

filtered_data
```

```python
#Q3
import numpy as np
import matplotlib.pyplot as plt

# Create a sample numerical dataset
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = np.sin(x)  # Sine function 

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='-', linewidth=2)

# Add a title, axis labels, and a legend
plt.title("Trend of a Numerical Dataset: Sine Wave", fontsize=16)
plt.xlabel("X-axis: Input Values", fontsize=12)
plt.ylabel("Y-axis: Sine Values", fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
#Q4
import pandas as pd
import numpy as np

# Create a DataFrame with random values and introduce some NaN values
np.random.seed(42)  
df = pd.DataFrame({
    'Column_A': np.random.randint(1, 100, 10),
    'Column_B': np.random.randint(1, 100, 10),
    'Column_C': np.random.randint(1, 100, 10)
})

# Introduce NaN values randomly
df.iloc[2, 1] = np.nan
df.iloc[5, 2] = np.nan

# (i) Replace NaN values with the mean of the respective column
df_filled = df.copy()
df_filled.fillna(df.mean(), inplace=True)

# (ii) Sort the DataFrame based on two columns: 'Column_A' and 'Column_B'
df_sorted = df_filled.sort_values(by=['Column_A', 'Column_B'])

df, df_filled, df_sorted
```

```python
#Q5
# Create two random matrices of compatible dimensions for dot product
matrix_a = np.random.randint(1, 10, (3, 3))  # 3x3 matrix
matrix_b = np.random.randint(1, 10, (3, 3))  # 3x3 matrix

# Calculate the dot product
dot_product = np.dot(matrix_a, matrix_b)

# Calculate the determinant of the resulting matrix
determinant = np.linalg.det(dot_product)

matrix_a, matrix_b, dot_product, determinant
```

```python
#Q6
import pandas as pd

# Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'C', 'A'],
    'Values': [10, 20, 30, 40, 50, 60, 70]
}
df = pd.DataFrame(data)

# Group by the 'Category' column and calculate the sum of 'Values' for each group
grouped_sum = df.groupby('Category')['Values'].sum()

print(grouped_sum)
```

```python
#Q7
import matplotlib.pyplot as plt
import numpy as np

# Create a numerical dataset
data = np.random.normal(loc=50, scale=10, size=1000)  # Normally distributed data

# Plot a histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, color='blue', edgecolor='black', alpha=0.7)

# Add title, labels, and grid lines
plt.title("Histogram of Numerical Dataset", fontsize=16)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#Q9
import pandas as pd

# Create the first DataFrame
data1 = {
    'ID': [1, 2, 3, 4],
    'Name': ['lakshansh', 'dyvanshu', 'ajay', 'kunal'],
    'Age': [21, 20, 19, 22]
}
df1 = pd.DataFrame(data1)

# Create the second DataFrame
data2 = {
    'ID': [3, 4, 5, 6],
    'Department': ['CHE', 'CE', 'MSE', 'SDS'],
    'CPI': [8.1, 8.5, 9.1, 8.4]
}
df2 = pd.DataFrame(data2)

# Perform the merge operation based on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='inner')  # Use 'inner' join for common IDs

print(merged_df)
```
