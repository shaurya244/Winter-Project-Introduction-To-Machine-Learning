```python
#this is the solution code for question number 4

import pandas as pd
import numpy as np


np.random.seed(42)  
data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10)
}
df = pd.DataFrame(data)


df.loc[2, 'A'] = np.nan
df.loc[5, 'B'] = np.nan
df.loc[7, 'C'] = np.nan


df_filled = df.fillna(df.mean())


sorted_df = df_filled.sort_values(by=['A', 'B'], ascending=[True, False])

# Display results
print("Original DataFrame with NaN values:")
print(df)
print("\nDataFrame after replacing NaN values with column means:")
print(df_filled)
print("\nDataFrame sorted by columns 'A' (ascending) and 'B' (descending):")
print(sorted_df)





```
