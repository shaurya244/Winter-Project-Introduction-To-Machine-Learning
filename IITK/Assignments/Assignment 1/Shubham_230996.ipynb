```python
import pandas as pd
import numpy as np
n=np.nan
data = {"A": [1,2,3,4,5,6,4,7,7,8],"B": [4,7,6,8,3,5,7,8,9,5],"C": [6,8,4,5,7,3,4,6,7,4]}
df = pd.DataFrame(data)

df['A']=df['A'].replace(2,n)
df['B']=df['B'].replace(6,n)
df['C']=df['C'].replace(4,n)

df.head(10)

df['A']=df['A'].replace(n,df['A'].mean())
df['B']=df['B'].replace(n,df['B'].mean())
df['C']=df['C'].replace(n,df['C'].mean())
df.head(10)
```
