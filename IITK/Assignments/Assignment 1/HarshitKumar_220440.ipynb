## Q1


```python
import numpy as np

# Create a 3D array with random numbers between 1 and 100
array_3d = np.random.randint(1, 101, size=(4, 4, 4))

# Calculate the mean and standard deviation
mean = np.mean(array_3d)
std_dev = np.std(array_3d)

# Perform z-score normalization
z_score_normalized = (array_3d - mean) / std_dev

print("Original 3D Array:\n", array_3d)
print("Z-Score Normalized Array:\n", z_score_normalized)
```

## Q4. Pandas DataFrame 


```python
# Create a DataFrame with random values and NaN
np.random.seed(0)
data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': [np.nan if i % 2 == 0 else i for i in range(10)]
}
df = pd.DataFrame(data)

# (i) Replace NaN values with the mean of the column
df['C'] = df['C'].fillna(df['C'].mean())

# (ii) Sort the DataFrame based on columns 'A' and 'B'
sorted_df = df.sort_values(by=['A', 'B'])

print("Original DataFrame:\n", df)
print("Sorted DataFrame:\n", sorted_df)
```

## Q5. Dot product and determinant


```python
# Create two matrices
matrix1 = np.random.randint(1, 10, size=(3, 3))
matrix2 = np.random.randint(1, 10, size=(3, 3))

# Calculate the dot product
dot_product = np.dot(matrix1, matrix2)

# Find the determinant of the resulting matrix
determinant = np.linalg.det(dot_product)

print("Matrix 1:\n", matrix1)
print("Matrix 2:\n", matrix2)
print("Dot Product:\n", dot_product)
print("Determinant of the Dot Product Matrix:", determinant)
```

## Q6. Grouping and aggregation with Pandas


```python
# Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'C', 'C', 'A'],
    'Values': [10, 20, 15, 25, 10, 30, 5]
}
df = pd.DataFrame(data)

# Group by 'Category' and calculate the sum of 'Values'
grouped_sum = df.groupby('Category')['Values'].sum()

print(grouped_sum)
```

## Q8. Slicing a 2D NumPy array


```python
# Create a 2D array
array_2d = np.random.randint(1, 20, size=(5, 5))

# Slice a sub-array (e.g., rows 1-3 and columns 2-4)
sub_array = array_2d[1:4, 2:5]

print("Original Array:\n", array_2d)
print("Sliced Sub-Array:\n", sub_array)
```

## Q9. Merge two DataFrames


```python
# Create two DataFrames
data1 = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
data2 = {'ID': [2, 3, 4], 'Score': [85, 90, 95]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Perform a merge on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='inner')

print(merged_df)

```
