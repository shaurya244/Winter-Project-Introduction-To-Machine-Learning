#### Using NumPy, create a 3D array of shape (4, 4, 4) filled with random numbers between 1 and 100. Find the z score normalisation of each element of the array.

```python
import numpy as np

arr = np.random.randint(1,101,(4,4,4))
print(arr)
import numpy as np
mean = np.mean(arr)
std = np.std(arr)
z_score = (arr-mean)/std
z_score
```

## Q3.

#### Using Matplotlib, plot a line graph showing the trend of a numerical dataset. Customize the plot with title, axis labels, and a legend.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
plt.figure(figsize = (8,8))
plt.xticks(np.arange(1, 30, 1))
plt.yticks(np.arange(0, 101, 5))
plt.plot(study_hours, marks, 'r-')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Marks vs Number of Study hours')
plt.legend(['Marks vs study hours'])
plt.show()
```

## Q5.

#### Using NumPy, calculate the dot product of two matrices and find the determinant of the resulting matrix.

```python
import numpy as np
a = np.array([[1, 2],[3, 4]])
b = np.array([[5, 6],[7, 8]])
print(a,"\n")
print(b,"\n")
c = a@b
print(c,"\n")
determinant = np.linalg.det(c)
print(int(determinant))
```
