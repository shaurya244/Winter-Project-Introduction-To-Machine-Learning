```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# Q1: Z-score normalization for a 3D array
array_3d = np.random.randint(1, 101, size=(4, 4, 4))
mean = np.mean(array_3d)
std_dev = np.std(array_3d)
z_score_normalized = (array_3d - mean) / std_dev
print("Original 3D Array:")
print(array_3d)
print("\nZ-score Normalized 3D Array:")
print(z_score_normalized)
```

```python
# Q2: Filter rows in Iris dataset where petal length > 3.0
iris = load_iris(as_frame=True).frame
filtered_iris = iris[iris['petal length (cm)'] > 3.0]
print("\nFiltered Iris Dataset:")
print(filtered_iris)
```

```python
# Q3: Line graph with Matplotlib
data = np.linspace(0, 10, 100)
square_data = data ** 2
plt.figure(figsize=(8, 6))
plt.plot(data, square_data, label="y = x^2", color='blue')
plt.title("Trend of Squared Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q4: DataFrame operations with random values
data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': [np.nan if i % 2 == 0 else np.random.rand() for i in range(10)]
}
df = pd.DataFrame(data)
print("\nOriginal DataFrame:")
print(df)

# Replace NaN values with column mean
df.fillna(df.mean(), inplace=True)
print("\nDataFrame after replacing NaN values:")
print(df)

# Sort DataFrame based on two columns
df.sort_values(by=['A', 'B'], inplace=True)
print("\nSorted DataFrame:")
print(df)
```

```python
# Q5: Dot product and determinant
matrix_a = np.random.randint(1, 10, (3, 3))
matrix_b = np.random.randint(1, 10, (3, 3))
dot_product = np.dot(matrix_a, matrix_b)
determinant = np.linalg.det(dot_product)
print("\nMatrix A:")
print(matrix_a)
print("\nMatrix B:")
print(matrix_b)
print("\nDot Product:")
print(dot_product)
print("\nDeterminant of Dot Product:", determinant)
```

```python
# Q6: Group DataFrame by a categorical column
categories = ['A', 'B', 'C', 'D']
data = {
    'Category': np.random.choice(categories, size=20),
    'Values': np.random.randint(1, 100, 20)
}
df_grouped = pd.DataFrame(data)
print("\nOriginal Grouped DataFrame:")
print(df_grouped)

group_sum = df_grouped.groupby('Category')['Values'].sum()
print("\nSum of Values by Category:")
print(group_sum)
```

```python
# Q7: Histogram with Matplotlib
data_histogram = np.random.normal(loc=50, scale=15, size=500)
plt.figure(figsize=(8, 6))
plt.hist(data_histogram, bins=20, color='purple', edgecolor='black')
plt.title("Distribution of Numerical Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
# Q8: Slicing a 2D array
array_2d = np.random.randint(1, 101, size=(6, 6))
sub_array = array_2d[2:5, 1:4]
print("\nOriginal 2D Array:")
print(array_2d)
print("\nSliced Sub-Array:")
print(sub_array)
```

```python
# Q9: Merge operation between two DataFrames
data1 = {
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
}
data2 = {
    'ID': [3, 4, 5, 6],
    'Score': [85, 90, 88, 92]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
merged_df = pd.merge(df1, df2, on='ID', how='inner')
print("\nMerged DataFrame:")
print(merged_df)
```

```python
# Q10: Scatter plot with Matplotlib
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='green', marker='o', label='Data Points')
plt.title("Scatter Plot of Random Data")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
