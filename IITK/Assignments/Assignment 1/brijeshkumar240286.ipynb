```python
#ques4**

import pandas as pd
import numpy as np

np.random.seed(43)

data = {
    'P': np.random.rand(21),
    'U': np.random.rand(21),
    'J': np.random.rand(21),
    'L': np.random.rand(21)
}


for column in data:
    data[column][np.random.choice(data[column].shape[0], size=4, replace=False)] = np.nan

x = pd.DataFrame(data)
print(x)



x.fillna(x.mean(), inplace=True)
print(x)

x_New = x.sort_values(by=['P', 'U', 'J', 'L'])
print("\n\n\n")
print(x_New)

```

```python
#ques5**


import numpy as np

X = np.array([[20, 8],
              [21, 10]])

Y = np.array([[20, 20],
              [21, 20]])

ans = np.dot(X, Y)

print(ans)


determinant = np.linalg.det(ans)
print("\n")
print(determinant)


```
