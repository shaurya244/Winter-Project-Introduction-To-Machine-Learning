Using NumPy, create a 3D array of shape (4, 4, 4) filled with random numbers between 1 and 100. Find the z score normalisation of each element of the array.

```python
import numpy as np 
```

```python
arr = np.random.randint(1,101,size=(4,4,4))
m = arr.mean()
d= arr.std()
arr = arr - m
arr = arr//d 
print(arr)
```

Using Pandas, filter the rows in the Iris dataset where the petal length is greater than 3.0, and display the filtered data 

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data,columns= iris.feature_names)
print(df)
filtered_df = df[df['petal length (cm)']>3] 
print(filtered_df)
```

Create a Pandas DataFrame with random values with at least 3 columns and perform the following operations: (i) Replace NaN values with the mean of the column. (ii) Sort the DataFrame based on any two columns.

```python
column_names = ['one','two','three']
random_num = np.random.randint(1,100,size=(3,3))
random_df= pd.DataFrame(random_num,columns=column_names)
nan_ind = np.random.choice([True,False],size=random_num.shape,p=[0.2,0.8])
random_df=random_df.mask(nan_ind)

print(random_df)
df_new = random_df.replace(np.nan,random_num.mean())
print(df_new)
```

Using NumPy, calculate the dot product of two matrices and find the determinant of the resulting matrix.

```python
arr1 = np.random.randint(1,10,size=(3,3))
arr2= np.random.randint(1,10 ,size=(3,3))
arr3 = arr1.dot(arr2)
determinant = np.linalg.det(arr3)
print(determinant)
```

With Pandas, group a DataFrame by a categorical column and calculate the sum of another numerical column for each group.

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris.data,columns= iris.feature_names)
print(df)
df.groupby('petal width (cm)').sum()['petal length (cm)']
```

Using Pandas, perform a merge operation between two DataFrames based on a common column for dataset of your choice.

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1,2,3], 'Score': [85, 90, 95]})
new=pd.merge(df1,df2,on='ID',how="left")
print(new)
```
