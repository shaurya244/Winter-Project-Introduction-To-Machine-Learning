**Q1.** Using NumPy, create a 3D array of shape (4, 4, 4) filled with random numbers between 1 and 100. Find the z score normalisation of each element of the array.

```python
import numpy as np
array_3d = np.random.randint(1, 101, size=(4, 4, 4))
print("Original Array:\n", array_3d)

mean = np.mean(array_3d)
std = np.std(array_3d)

z_score_normalized = (array_3d - mean) / std
print("\nZ-score Normalized Array:\n", z_score_normalized)
```

**Q2.** Using Pandas, filter the rows in the Iris dataset where the petal length is greater than 3.0, and display the filtered data.

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = iris.frame

filtered_data = df[df['petal length (cm)'] > 3.0]

print("Filtered data : ",filtered_data)
```

**Q3.** Using Matplotlib, plot a line graph showing the trend of a numerical dataset. Customize the plot with title, axis labels, and a legend.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 6))

plt.plot(x, y1, label='Sine Wave', color='b', linestyle='-', marker='o')
plt.plot(x, y2, label='Cosine Wave', color='r', linestyle='--', marker='x')

plt.title("Trend of Sine and Cosine Waves")
plt.xlabel("X-axis (radians)")
plt.ylabel("Y-axis (amplitude)")

plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper right')

plt.show()
```

**Q4.** Create a Pandas DataFrame with random values with at least 3 columns and perform the following operations: (i) Replace NaN values with the mean of the column. (ii) Sort the DataFrame based on any two columns.

```python
import pandas as pd
import numpy as np

np.random.seed(42)
data = {
    'A': np.random.randn(10),
    'B': np.random.randn(10),
    'C': np.random.choice([np.random.randn(), np.nan], 10)
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

df.fillna(df.mean(numeric_only=True), inplace=True)
print("\nDataFrame after Replacing NaNs with Column Means:\n", df)

sorted_df = df.sort_values(by=['A', 'B'], ascending=[True, False])
print("\nDataFrame Sorted by Columns A (Ascending) and B (Descending):\n", sorted_df)
```

**Q5.** Using NumPy, calculate the dot product of two matrices and find the determinant of the resulting matrix.

```python
import numpy as np

A = np.array([[2, 3], [4, 5]])
B = np.array([[1, 2], [3, 4]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

result = np.dot(A, B)
print("\nDot Product of A and B:\n", result)

determinant = np.linalg.det(result)
print("\nDeterminant of the Resulting Matrix:", round(determinant, 2))
```

**Q6.** With Pandas, group a DataFrame by a categorical column and calculate the sum of another numerical column for each group.

```python
import pandas as pd

data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
    'Values': [10, 20, 15, 30, 25, 5, 40, 35]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

grouped_sum = df.groupby('Category')['Values'].sum()

print("\nSum of Values by Category:\n", grouped_sum)
```

**Q7.** Using Matplotlib, create a histogram to show the distribution of a numerical dataset. Customize the number of bins and add grid lines for better readability.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=50, scale=10, size=1000)

plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)

plt.title("Distribution of Numerical Data")
plt.xlabel("Data Values")
plt.ylabel("Frequency")

plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

**Q8.** Using NumPy, create a 2D array and perform a slicing operation to extract a specific sub-array.

**Q9.** Using Pandas, perform a merge operation between two DataFrames based on a common column for dataset of your choice.

```python
import pandas as pd

df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Department': ['HR', 'Finance', 'IT', 'Marketing']
})

df2 = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4],
    'Salary': [50000, 60000, 55000, 65000]
})

print("DataFrame 1 (Employees):\n", df1)
print("\nDataFrame 2 (Salaries):\n", df2)

merged_df = pd.merge(df1, df2, on='EmployeeID')

print("\nMerged DataFrame:\n", merged_df)
```

**Q10.** Using Matplotlib, create a scatter plot comparing two numerical variables in the dataset of your choice. Customize the plot by changing the color and shape of the data points.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50) * 100 
y = np.random.rand(50) * 100 

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c='red', marker='^', edgecolors='black', s=100, label='Data Points')

plt.title('Scatter Plot of Two Numerical Variables', fontsize=14)
plt.xlabel('Variable X', fontsize=12)
plt.ylabel('Variable Y', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```
