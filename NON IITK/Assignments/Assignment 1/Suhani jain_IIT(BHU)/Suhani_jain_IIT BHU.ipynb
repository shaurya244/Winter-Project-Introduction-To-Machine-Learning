Q1 -Using NumPy, create a 3D array of shape (4, 4, 4) filled with random numbers between 1 and 100. Find the z score normalisation of each element of the array.

```python
ar = np.random.randint(1, 101 , size = (4 , 4, 4))
print(ar)
```

```python
# calculate the standard deviation and mean 
mean = np.mean(ar)
std = np.std(ar)

# Perform Z-score normalization
z_score_normalized_array = (ar - mean) / std
print(z_score_normalized_array)
```

Q4.Create a Pandas DataFrame with random values with at least 3 columns and perform the following operations: (i) Replace NaN values with the mean of the column. (ii) Sort the DataFrame based on any two columns.

```python
# Create a DataFrame with random values and some NaN values
np.random.seed(42)  # For reproducibility
data = {
     "c1" : np.random.choice([np.nan, *np.random.randint(1, 100, 10)], 10),
    "c2" : np.random.randint(1 , 100 , 10),
    "c3": np.random.randint(1, 100, 10),
}
df = pd.DataFrame(data)

# (i) Replace NaN values with the mean of the column
df_filled = df.copy()
df_filled["c1"].fillna(df_filled["c1"].mean(), inplace=True)

# (ii) Sort the DataFrame based on Column2 and Column3 in ascending order
df_sorted = df_filled.sort_values(by=["c2", "c3"], ascending=[True, True])

df, df_filled, df_sorted
```

Q7. Using Matplotlib, create a histogram to show the distribution of a numerical dataset. Customize the number of bins and add grid lines for better readability.

```python
#finding numerical columns
num_col = df.drop(['Id' , 'Species'] , axis = 1)
# Plot separate histograms for each column
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()  # Flatten to easily loop through

for i , column in enumerate(num_col):
    axes[i].hist(df[column], bins=10, color='blue', alpha=0.7, edgecolor='black')
    axes[i].set_title(f"Distribution of {column}")
    axes[i].set_xlabel("Value Ranges")
    axes[i].set_ylabel("Frequency")
    axes[i].grid(alpha=0.3)

# Adjust layout
plt.tight_layout()
plt.show()
```

Q9. Using Pandas, perform a merge operation between two DataFrames based on a common column for dataset of your choice.

```python
# Create a DataFrames with random values 
np.random.seed(42)  # For reproducibility
data_1 = {
     "c1" : np.random.randint(1 , 15 , 10),
    "c2" : np.random.randint(1 , 100 , 10),
    "c3": np.random.randint(1, 200, 10),
}
df_1 = pd.DataFrame(data_1)

np.random.seed(42)
data_2 = {
     "c4" : np.random.randint(1 , 100 , 10),
    "c5" : np.random.randint(1 , 10 , 10),
    "c3": np.random.randint(1, 200, 10),
}
df_2 = pd.DataFrame(data_2)

#merge two dataframes based on common column(c3)
merge_df = pd.merge(df_1 , df_2 , on = 'c3' , how = 'inner')
merge_df
```
