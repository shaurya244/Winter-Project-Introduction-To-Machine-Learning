```python
# Q1 Using NumPy, create a 3D array of shape (4, 4, 4) filled with random numbers between 1 and 100. Find the z score normalisation of each element of the array.
import numpy as np

array = np.random.randint(1, 101, size=(4, 4, 4))
mean = np.mean(array)
std_dev = np.std(array)
z_score_array = (array - mean) / std_dev
print(array, z_score_array)
```

```python
#. Q2 Using Pandas, filter the rows in the Iris dataset where the petal length is greater than 3.0, and display the filtered data
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
filtered_df = df[df['petal length (cm)'] > 3.0]

print(filtered_df)
```

```python
#Q4 Create a Pandas DataFrame with random values with at least 3 columns and perform the following operations: (i) Replace NaN values with the mean of the column. (ii) Sort the DataFrame based on any two columns.

import pandas as pd
import numpy as np
np.random.seed(42)
data = {
    'A': np.random.randint(1, 10, size=10),
    'B': np.random.rand(10),
    'C': np.random.randint(1, 100, size=10)
}

df = pd.DataFrame(data)
nan_indices = np.random.choice(df.index, size=3, replace=False)
df.loc[nan_indices, 'B'] = np.nan  # Set NaN values in column 'B'
df['B'].fillna(df['B'].mean(), inplace=True)
df_sorted = df.sort_values(by=['A', 'C'])

print("Original DataFrame with NaN values:")
print(df)

print("\nSorted DataFrame:")
print(df_sorted)
```

```python
#Q5 Using NumPy, calculate the dot product of two matrices and find the determinant of the resulting matrix
import numpy as np
A = np.array([[3, 5], [6, 8]])
B = np.array([[2, 6], [1, 9]])
C = np.dot(A, B)
det_C = np.linalg.det(C)
print("Dot product of A and B:\n", C)
print("Determinant of the resulting matrix:", det_C)
```

```python
#Q6 . With Pandas, group a DataFrame by a categorical column and calculate the sum of another numerical column for each group.
import pandas as pd
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Value': [10, 20, 30, 40, 50, 60]
}

df = pd.DataFrame(data)
grouped = df.groupby('Category')['Value'].sum().reset_index()

print(grouped)

```

```python
# Q7 Using Matplotlib, create a histogram to show the distribution of a numerical dataset. Customize the number of bins and add grid lines for better readability
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)  # 1000 random data points from a normal distribution
plt.hist(data, bins=30, edgecolor='black')  # Customizing the number of bins
plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

```python
# Q9 Using Pandas, perform a merge operation between two DataFrames based on a common column for dataset of your choice
import pandas as pd
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Alice', 'Bob', 'Eve', 'Charlie'],
    'department_id': [101, 102, 103, 101, 102]
})
departments = pd.DataFrame({
    'department_id': [101, 102, 103],
    'department_name': ['HR', 'Finance', 'Engineering']
})
merged_data = pd.merge(employees, departments, on='department_id')
print(merged_data)
```

```python
#Q10 . Using Matplotlib, create a scatter plot comparing two numerical variables in the dataset of your choice. Customize the plot by changing the color and shape of the data points.
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris()
sepal_length = iris.data[:, 0]  # Sepal length (first column)
sepal_width = iris.data[:, 1]   # Sepal width (second column)
plt.figure(figsize=(8, 6))
plt.scatter(sepal_length, sepal_width, color='green', marker='^', linewidths=1, edgecolor='black')
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()
```
