```python
data1= {'S_no.' : [1,2,3],
       'name' : ['apple', 'mango', 'orange']}

df=pd.DataFrame(data1)
print(df)
print('-'*20)

data2= {'S_no.' : [1,2,3],
        'color' : ['red', 'yellow', 'ornge']}

df2=pd.DataFrame(data2)
print(df2)
print('-'*20)

merged = pd.merge(df, df2, on="S_no.", how="inner")

print(merged)
```

```python
x1 = np.random.randn(1000)
y1 = np.random.randn(1000)

x2 = np.random.randn(1000) + 2
y2 = np.random.randn(1000) + 2

plt.scatter(x1,y1, c = 'red',  alpha = .7, label = 'Group - 1')

plt.scatter(x2,y2, c = 'blue', alpha = .7, label = 'Group - 2')

plt.title('Scatter Plot with Multiple Groups')
plt.legend()
plt.show()
```
