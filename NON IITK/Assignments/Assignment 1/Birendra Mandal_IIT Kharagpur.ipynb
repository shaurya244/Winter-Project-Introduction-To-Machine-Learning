# Q1

```python
import numpy as np

# Create a 3D array of shape (4, 4, 4) with random integers between 1 and 100
array = np.random.randint(1, 101, size=(4, 4, 4))

# Calculate the mean and standard deviation of the array
mean = np.mean(array)
std_dev = np.std(array)

# Perform z-score normalization
z_score_normalized_array = (array - mean) / std_dev

print("Original array:\n", array)
print("\nZ-score normalized array:\n", z_score_normalized_array)
```

# Q2

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Filter rows where petal length is greater than 3.0
filtered_df = df[df['petal length (cm)'] > 3.0]

# Display the filtered data
print(filtered_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.randn(100) * 0.5

# Plot the line graph
plt.plot(x, y, label='Sine Wave with Noise')

# Customize the plot
plt.title('Sine Wave with Noise')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Display the plot
plt.show()
```

# Q4

```python
import pandas as pd
import numpy as np

# Create a DataFrame with random values
data = {'Column1': np.random.randn(10),
        'Column2': np.random.randint(1, 10, 10),
        'Column3': np.random.uniform(10, 20, 10)}
df = pd.DataFrame(data)

# Introduce some NaN values
df.iloc[2, 0] = np.nan
df.iloc[5, 2] = np.nan

# Replace NaN values with the mean of the column
df.fillna(df.mean(), inplace=True)

# Sort the DataFrame based on two columns
df.sort_values(by=['Column1', 'Column2'], ascending=[False, True], inplace=True)

print(df)
```

# Q5

```python
import numpy as np

# Create two random matrices
matrix1 = np.random.randint(1, 10, size=(3, 3))
matrix2 = np.random.randint(1, 10, size=(3, 3))

# Calculate the dot product
dot_product_matrix = np.dot(matrix1, matrix2)

# Calculate the determinant of the dot product matrix
determinant = np.linalg.det(dot_product_matrix)

print("Matrix 1:\n", matrix1)
print("\nMatrix 2:\n", matrix2)
print("\nDot Product Matrix:\n", dot_product_matrix)
print("\nDeterminant of the Dot Product Matrix:", determinant)
```

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {'Category': ['A', 'B', 'A', 'C', 'B', 'A'],
        'Value': np.random.randint(1, 10, 6)}
df = pd.DataFrame(data)

# Group the DataFrame by 'Category' and calculate the sum of 'Value'
grouped_df = df.groupby('Category')['Value'].sum()

print(grouped_df)
```

# Q9

```python
import pandas as pd

# Create sample DataFrames
products = pd.DataFrame({'product_id': [1, 2, 3, 4],
                          'product_name': ['Product A', 'Product B', 'Product C', 'Product D'],
                          'category': ['Electronics', 'Clothing', 'Electronics', 'Books']})

sales = pd.DataFrame({'product_id': [1, 2, 3],
                       'sales_quantity': [100, 50, 200],
                       'sales_revenue': [10000, 5000, 25000]})

# Merge the DataFrames based on the 'product_id' column
merged_df = pd.merge(products, sales, on='product_id')

print(merged_df)
```
