# Question 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import det
```

```python
array_3d = np.random.randint(1, 101, size=(4, 4, 4))
mean = np.mean(array_3d)
std_dev = np.std(array_3d)
z_score_normalized = (array_3d - mean) / std_dev
print(array_3d)
print(z_score_normalized)
```

# Question 2

```python
iris_column_names = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]
iris_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None, names=iris_column_names + ["class"])
iris_df_filtered = iris_df[iris_df["petal length (cm)"] > 3.0]

print(iris_df_filtered)
```

# Question 4

```python
data = {
    'A': [np.random.rand() for _ in range(5)],
    'B': [np.random.rand() if i % 2 == 0 else np.nan for i in range(5)],
    'C': [np.random.randint(1, 100) for _ in range(5)]
}
df = pd.DataFrame(data)
df['B'].fillna(df['B'].mean(), inplace=True)
df_sorted = df.sort_values(by=['A', 'C'])

print(df_sorted)
```

# Question 5

```python
matrix1 = np.random.randint(1, 10, (3, 3))
matrix2 = np.random.randint(1, 10, (3, 3))
dot_product = np.dot(matrix1, matrix2)
determinant = det(dot_product)

print(matrix1)
print(matrix2)
print(dot_product)
print(determinant)
```

# Question 6

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'C'],
    'Values': [10, 20, 15, 25, 30]
}
df_group = pd.DataFrame(data)
grouped_sum = df_group.groupby('Category')['Values'].sum()

print(grouped_sum)
```

# Question 7

```python
numerical_data = np.random.normal(50, 10, 1000)
plt.figure(figsize=(8, 6))
plt.hist(numerical_data, bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Numerical Dataset", fontsize=14)
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(True)
plt.show()
```

# Question 9

```python
data1 = {
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
}
data2 = {
    'ID': [2, 3, 4],
    'Score': [85, 90, 95]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
merged_df = pd.merge(df1, df2, on='ID', how='inner')

print(merged_df)
```
