Q1. Using NumPy, create a 3D array of shape (4, 4, 4) filled with random numbers between 1 and 100. Find the z score normalisation of each element of the array.

Q4.Create a Pandas DataFrame with random values with at least 3 columns and perform the following operations: (i) Replace NaN values with the mean of the column. (ii) Sort the DataFrame based on any two columns.

```python

data = {
    "num1": np.random.randint(10, 100, 10),
    "num2": np.random.choice([np.nan, *np.random.randint(10, 100, 9)], 10),
    "num3": np.random.choice([np.nan, *np.random.randint(10, 100, 9)], 10),
}
df = pd.DataFrame(data)
df
```

```python
df.fillna(df.mean(), inplace = True)
df
```

```python
sorted_df = df.sort_values(by=["num1", "num2"])
sorted_df
```

Q6. With Pandas, group a DataFrame by a categorical column and calculate the sum of another numerical column for each group.

```python
data = {
    'Category': np.random.randint(10, 20, 10),
    'Values': np.random.randint(10, 100, 10)
}
df = pd.DataFrame(data)
grouped = df.groupby('Category')
grouped_sum = grouped['Values'].sum()
print(grouped_sum)
```

Q9. Using Pandas, perform a merge operation between two DataFrames based on a common column for dataset of your choice.

```python
df1 = pd.DataFrame({
    'Product_id' : [1001, 1002, 1003, 1004],
    'Product_name' : ['Bluetooth', 'Hard drive', 'Monitor', 'Keyboard'],
    'MRP' : [1000, 2000, 4500, 2500]
})
df2 = pd.DataFrame({
    'Product_id' : [1001, 1002, 1003, 1004],
    'Total_quantity' : [80, 95, 45, 85],
    'Quantity_sold' : [60, 50, 15, 20]
})
merged_df = pd.merge(df1, df2, on='Product_id', how='inner')
print(merged_df)
```
