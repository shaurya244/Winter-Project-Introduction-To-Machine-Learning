Q4.Create a Pandas DataFrame with random values with at least 3 columns and perform the following operations: (i) Replace NaN values with the mean of the column. (ii) Sort the DataFrame based on any two columns.

```python
np.random.seed(42)
df = pd.DataFrame({
    'A': np.random.randint(1, 100, 10),
    'B': np.random.rand(10) * 100,
    'C': np.random.choice([np.nan, 10, 20, 30, 40], size=10)
})
```

```python
df.loc[2, 'A'] = np.nan
df.loc[5, 'B'] = np.nan
```

```python
df
```

```python
df['A']=df['A'].fillna(df['A'].mean())
df['B']=df['B'].fillna(df['B'].mean())
df['C']=df['C'].fillna(df['C'].mean())
```

```python
df
```

```python
df_sorted = df.sort_values(by=['A', 'B'], ascending=[True, False])
```

```python
df_sorted
```

Q9. Using Pandas, perform a merge operation between two DataFrames based on a common column for dataset of your choice.

```python
students = pd.DataFrame({
    'ID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [20, 22, 21, 23],
    'Gender': ['Female', 'Male', 'Male', 'Male']
})
scores = pd.DataFrame({
    'ID': [101, 102, 103, 105],
    'Subject': ['Math', 'Science', 'History', 'English'],
    'Score': [85, 90, 75, 88]
})
```

```python
stdscores=pd.merge(students,scores,on='ID')
```

Q10. Using Matplotlib, create a scatter plot comparing two numerical variables in the dataset of your choice. Customize the plot by changing the color and shape of the data points.

```python
data = pd.DataFrame({
    'Age': [20, 22, 21, 23, 19, 25, 24, 30],
    'Score': [85, 90, 75, 88, 95, 70, 80, 60]
})
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(data['Age'], data['Score'],
            color='red',
            s=100,
            edgecolor='black')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('Scatter Plot of Age vs Score', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
