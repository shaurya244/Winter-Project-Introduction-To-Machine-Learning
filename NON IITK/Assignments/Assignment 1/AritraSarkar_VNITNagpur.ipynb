```python

np.random.seed(42)
data={
    'column1':np.random.randint(1,100,size=10),
    'column2':[np.nan if i%3 ==0 else np.random.rand()*100 for i in range(10)],
    'column3':np.random.randint(1,50,size=10)
}

df=pd.DataFrame(data)
print (df)

df1=df.fillna(df.mean(), inplace=True)
print(df1)

new_df=df.sort_values(by=['column1','column2'],ascending=[True,False])

print(new_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
x=np.random.rand(100)*50
y= x+np.random.rand(100)*10

plt.figure(figsize=(8,6))
plt.scatter(x,y,c='purple',marker='^',alpha=0.8,edgecolor='black',s=50)

plt.title('scatter plot of data',fontsize=16)
plt.xlabel('variable x',fontsize=14)
plt.ylabel('variable y',fontsize=14)

plt.show()
```
