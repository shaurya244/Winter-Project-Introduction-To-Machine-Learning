```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_excel('ENB2012_data.xlsx')
data.head()
```

```python
columns = ["Relative_Compactness", "Surface_Area", "Wall_Area", "Roof_Area", "Overall_Height",
           "Orientation", "Glazing_Area", "Glazing_Area_Distribution", "Heating_Load", "Cooling_Load"]
data.columns = columns
data.isnull().sum()
```

```python
data.corr()
```

```python
#correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

Splitting the dataset in 80:20 ratio

```python
data.info()
```

```python
x = data.loc[:,'Relative_Compactness':'Glazing_Area_Distribution']
y = data.loc[:,['Heating_Load','Cooling_Load']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state = 31)
```

Training a linear regression model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

Training a lasso regression model and ridge regression model

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
alpha = 0.11
lasso = Lasso(alpha = alpha)
lasso.fit(x_train,y_train)
```

```python
lasso.coef_
```

```python
lasso.intercept_
```

```python
ridge = Ridge(alpha=alpha)
ridge.fit(x_train,y_train)
```

```python
ridge.coef_
```

```python
ridge.intercept_
```

```python
from sklearn.metrics import mean_squared_error,r2_score
predicted_value_lr = lr.predict(x_test)
predicted_value_lasso = lasso.predict(x_test)
predicted_value_ridge = ridge.predict(x_test)
```

```python
cost_lr = mean_squared_error(y_test, predicted_value_lr)
cost_lasso = mean_squared_error(y_test, predicted_value_lasso)
cost_ridge = mean_squared_error(y_test, predicted_value_ridge)
```

Now for r2 score

```python
cost_lr = r2_score(y_test, predicted_value_lr)
cost_lasso = r2_score(y_test, predicted_value_lasso)
cost_ridge = r2_score(y_test, predicted_value_ridge)
```
